```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("hr_employee_data (1).xlsx")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
## Clarifying the etiquette name of data for left and stayed in the company
df["left"] = df["left"].map({0: "Stayed", 1: "Left"})
```

```python
sns.countplot(x="left", data=df)
plt.title("Left vs stayed")
plt.show()
```

```python
## Comparing satisfaction level based on left and stayed.
sns.boxplot(x="left", y="satisfaction_level", data=df)
plt.title("Satisfaction based on left or stayed")
plt.show()
```

```python
## How many working hours on the people that left and stayed

sns.boxplot(x="left", y="average_montly_hours", data=df)

plt.title("Working hours vs stayed and left")
plt.show()
```

```python
## Visualizing amout of projects per left and stayed.
sns.histplot(data=df, x="number_project", hue="left", multiple="stack")

plt.title("Amount of project vs left and stayed")
plt.show()
```

```python
## Stats of avg monthly hours and number of project on left and stayed 
df.groupby("left")[["satisfaction_level", "average_montly_hours", "number_project"]].mean()
```
