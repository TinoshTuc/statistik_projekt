```python
## Make it with all variables 
X_full, y_full = load_iris(return_X_y=True, as_frame=True)

X_train_full_f, X_test_f, y_train_full_f, y_test_f = train_test_split(X_full, y_full, test_size=0.2, random_state=40)

X_train_f, X_val_f, y_train_f, y_val_f = train_test_split(X_train_full_f, y_train_full_f, test_size=0.3, random_state=36)
```

```python
# Train improved model
model_f = LogisticRegression(max_iter=200)
model_f.fit(X_train_f, y_train_f)
```

```python
# Validation
y_val_pred_f = model_f.predict(X_val_f)
print("model f validation")
print(classification_report(y_val_f, y_val_pred_f))
```

```python
# Test
y_test_pred_f = model_f.predict(X_test_f)
print("model f test")
print(classification_report(y_test_f, y_test_pred_f))
```
